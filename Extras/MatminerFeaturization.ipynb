```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tabulate as tabulate

df = pd.read_csv(r"C:\Users\joeya\Documents\Fokwa Group\ML_tutorial\intermetallics\intermetallics_Fulldataset_cleaned2.csv")

print(df.shape)
display(df)
```

```python
from matminer.featurizers.conversions import StrToComposition

stc = StrToComposition() 

if __name__ == '__main__':
    df = stc.featurize_dataframe(df, 'formula', ignore_errors=(True))

display(df)
```

```python
from matminer.featurizers.composition.element import TMetalFraction
tmf = TMetalFraction()

df = tmf.featurize_dataframe(df, 'composition', ignore_errors=(True))
```

```python
display(df)

from matminer.featurizers.composition.element import ElementFraction

ef = ElementFraction()
df = ef.featurize_dataframe(df, 'composition', ignore_errors=(True))

df.to_csv(r"C:\Users\joeya\Documents\Fokwa Group\ML_tutorial\intermetallics\intermetallics_Fulldataset_featurized_allElements2.csv", index=False)
```

```python
#%%
#make a new column which is an addition of the 3d & 5d element fractions in each composition

d3_frac = df['Cr'] + df['Mn'] + df['Fe'] + df['Co'] + df['Ni']
df['3d_frac'] = d3_frac


d5_frac = df['Hf'] + df['Ta'] + df['W'] + df['Re'] + df['Os'] + df['Ir'] + df['Pt'] + df['Au'] + df['Hg']
df['5d_frac'] = d5_frac

d4_frac = df['Y'] + df['Zr'] + df['Nb'] + df['Mo'] + df['Ru'] + df['Rh'] + df['Pd'] + df['Ag'] + df['Cd']
df['4d_frac'] = d4_frac

df['SumMnFe'] = df['Mn'] + df['Fe']
df['SumMnCr'] = df['Mn'] + df['Cr']
df['SumMnCo'] = df['Mn'] + df['Co']
df['SumMnNi'] = df['Mn'] + df['Ni']

df['SumFeNi'] = df['Fe'] + df['Ni']
df['SumFeCr'] = df['Fe'] + df['Cr']
df['SumFeCo'] = df['Fe'] + df['Co']

df['SumCrNi'] = df['Cr'] + df['Ni']
df['SumCrNi'] = df['Cr'] + df['Co']

df['SumCoNi'] = df['Co'] + df['Ni']

df['SumMnFeCo']= df['Mn'] + df['Fe'] +df['Co']
df['SumMnFeNi']= df['Mn'] + df['Fe'] +df['Ni']
df['SumMnFeCr']= df['Mn'] + df['Fe'] +df['Cr']
df['SumCrMnCo']= df['Cr'] + df['Mn'] +df['Co']
df['SumCrFeCo']= df['Cr'] + df['Fe'] +df['Co']
df['SumCrMnNi']= df['Cr'] + df['Mn'] +df['Ni']
df['SumCrFeNi']= df['Cr'] + df['Fe'] +df['Ni']
df['SumCrCoNi']= df['Cr'] + df['Co'] +df['Ni']
df['SumFeNiCo']= df['Fe'] + df['Ni'] +df['Co']
df['SumFeNiCo']= df['Fe'] + df['Ni'] +df['Co']


#To drop a column df=df.drop('SumFeMn', axis=1)

print(df.shape)
display(df)
```

```python
#Try to get amount of each 3d elements and plot it.

d3Elements_dict ={'Cr':0, 'Mn':0, 'Fe':0, 'Co':0, 'Ni':0}
for ind in df.index.values:
    entry = df.loc[ind]
    if entry['Cr'] > 0.000:
        d3Elements_dict['Cr'] += 1
    elif entry['Mn'] > 0.000:
        d3Elements_dict['Mn'] += 1
    elif entry['Fe'] > 0.000:
        d3Elements_dict['Fe'] += 1
    elif entry['Co'] > 0.000:
        d3Elements_dict['Co'] += 1
    elif entry['Ni'] > 0.000:
        d3Elements_dict['Ni'] += 1

plt.rcParams.update({'font.size': 25})
plt.figure(figsize=(12,12))
plt.bar(range(len(d3Elements_dict)),d3Elements_dict.values(), align='center')
plt.xticks(range(len(d3Elements_dict)), [str(key) for key in list(d3Elements_dict.keys())])
plt.title('Distribution of 3d elements within dataset')

plt.savefig(r'C:\Users\joeya\Documents\Fokwa Group\ML_tutorial\intermetallics\figures\3dElements_transparent.png', 
            bbox_inches='tight', dpi=600, transparent = True)
plt.savefig(r'C:\Users\joeya\Documents\Fokwa Group\ML_tutorial\intermetallics\figures\3dElements.png', 
            bbox_inches='tight', dpi=600)

    
```
